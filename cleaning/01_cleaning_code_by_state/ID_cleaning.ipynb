```python
import pandas as pd
import numpy as np
```

```python
abbr = 'ID'
```

Read in federal level data

```python
fiscal = pd.read_sas('../../data/fiscal2018', format = 'sas7bdat', encoding='iso-8859-1')
```

Generate list of districts in the state in the federal data

```python
state_fiscal = fiscal[(fiscal['STABBR'] == abbr) & (fiscal['GSHI'] == '12')]
```

```python
len(state_fiscal)
```

```python
state_fiscal.head()
```

Read in state level data

```python
state_grads = pd.read_excel('../../data/state_data_raw/idaho2018.xlsx', sheet_name='2018 District 4Year CohortGradR')
```

```python
state_grads
```

Reset columns

```python
state_grads.columns = state_grads.loc[1]
state_grads = state_grads.loc[2:]
```

```python
state_grads
```

Drop and rename columns.

```python
state_grads = state_grads[(state_grads['Population'] == 'All Students')]
```

```python
state_grads
```

Select and rename columns.

```python
state_grads = state_grads[['DistrictName', 'Cohort', 'Rate']]
```

```python
state_grads.columns = ['District Name', 'Total', 'Graduation Rate']
```

```python
state_grads
```

```python
state_grads.info()
```

Convert data types.

```python
state_grads['Graduation Rate'] = state_grads['Graduation Rate'].astype(str).str.replace('%', '')
```

```python
state_grads['Graduation Rate'] = pd.to_numeric(state_grads['Graduation Rate']) / 100
```

Check for matches and non-matches in the two lists

```python
Matches = [name for name in list(state_grads['District Name']) if name in list(state_fiscal['NAME'])]
Matches.sort()
len(Matches)
```

```python
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' INC.', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' SCHOOL', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' DISTRICT', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' CHARTER', '')

state_grads['District Name'] = state_grads['District Name'].str.replace(' SCHOOL', '')
state_grads['District Name'] = state_grads['District Name'].str.replace(' DISTRICT', '')
state_grads['District Name'] = state_grads['District Name'].str.replace(' CHARTER', '')
```

```python
A = [name for name in list(state_grads['District Name']) if name not in list(state_fiscal['NAME'])]
A.sort()
A
```

```python
B = [name for name in list(state_fiscal['NAME']) if name not in list(state_grads['District Name'])]
B.sort()
B
```

Replace the names I can find matches for.

```python
state_grads_rename = {
    'COMPASS PUBLIC' : 'COMPASS',
    'EMMETT INDEPENDENT' : 'EMMETT INDEPENDENT DIST',
    #'IDAHO BUREAU OF EDUCATIONAL SERVICES FOR THE DEAF AND THE BL',
    'IDAHO COLLEGE AND CAREER READINESS ACADEMY' : 'IDAHO COLLEGE & CAREER READINESS ACADEMY',
    #'IDAHO DEPARTMENT OF CORRECTIONS',
    #'IDAHO DEPARTMENT OF JUVENILE CORRECTIONS',
    #'IDAHO VIRTUAL EDUCATION PARTNERS',
    'INSPIRE ACADEMICS' : 'INSPIRE VIRTUAL',
    'ISUCCEED VIRTUAL HIGH' : 'iSUCCEED VIRTUAL HIGH',
    'JEFFERSON COUNTY JOINT' : 'JEFFERSON COUNTY JT',
    #'KTEC - Kootenai Tech Ed Campus' : ,
    'MCCALL-DONNELLY JOINT' : 'MCCALL-DONNELLY JT.',
    'PATHWAYS IN EDUCATION - NAMPA' : 'PATHWAYS IN EDUCATION - NAMPA,',
    'SALMON RIVER JOINT' : 'SALMON RIVER JOINT DIST',
    #'SANDPOINT',
    "TAYLOR'S CROSSING PUBLIC" : 'TAYLORS CROSSING',
    'THE KOOTENAI BRIDGE ACADEMY' : 'KOOTENAI BRIDGE ACADEMY ',
    'THE SAGE INTERNATIONAL OF BOISE A PUBLIC SCH' : 'SAGE INTERNATIONAL OF BOISE',
    'WHITEPINE JOINT' : 'WHITEPINE JT'
}
```

```python
state_fiscal = state_fiscal.replace(state_grads_rename)
```

Merge federal and state data, keeping only matches between the two. 

```python
state_grads_merged = pd.merge(state_fiscal, state_grads, how='inner', left_on='NAME', right_on='District Name')
```

Save cleaned data. 

```python
state_grads_merged.to_csv('../../data/state_data_merged/' + abbr + '.csv', index=False)
```
